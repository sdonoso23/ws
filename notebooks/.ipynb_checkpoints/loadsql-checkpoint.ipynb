```python
import sys
import os
sys.path.append(os.getcwd() + "/..")
import ws.scrapers as sc
import ws.jsontosql
import pandas as pd
import json
import pickle
import time
import sqlite3

conn = sqlite3.connect('../data/db.sqlite')
path = "../data/"
tournaments = ["La Liga","Bundesliga","Premier League","Serie A","Ligue 1"]
years = ["2017-2018","2016-2017","2015-2016","2014-2015","2013-2014","2012-2013","2011-2012","2010-2011","2009-2010"]
status = "append"
```

```python
for tournament in tournaments:
    for year in years:
        jsontosql.matchtosql(conn,path,tournament,year,status)
```

```python
for year in years:
    jsontosql.matchtosql(conn,path,"UEFA Champions League",year,status)
```

```python
matches = pd.read_sql("SELECT * FROM matches",conn,parse_dates = ["date"])
```
