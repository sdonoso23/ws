```python
for league in ["Bundesliga","Serie A"]:
    
    league_path = JSONPATH+league.lower().replace(" ","_")+"/"
    
    for season in os.listdir(league_path):
        
        season_path = league_path+season+"/"
        
        for match in filter(lambda x: "json" in x, os.listdir(season_path)):
            
            matchId = int(match.replace(".json",""))
            match_data = json.load(open(season_path+match,"r"))
            match_data.update({"matchId":matchId,"league":league,"season":season})
            coll.insert_one(match_data)
        print("League:",league,"Season:",season,"Done!")
```
